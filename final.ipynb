```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = "6dfd4f8f-cbf5-4917-a5eb-fd07f4403167.csv"  
```

```python
def isolate_column(df, column_name):
    if df is None or df.empty:
        print(f"error my friend, {column_name}")
        return None

    if column_name not in df.columns:
        print(f"error my friend, {column_name}")
        return None

    return df[[column_name]]
```

```python
traffic_df = pd.read_csv(csv_path)

overLimit = isolate_column(traffic_df, 'percent_over_limit') # isolating
neighborhood = isolate_column(traffic_df, 'neighborhood') # isolating
data = pd.concat([neighborhood, overLimit], axis=1) # cleaning
data = data.dropna() # more cleaning df
avg_percent = data.groupby('neighborhood')['percent_over_limit'].mean().sort_values(ascending=True) # getting average % over the speed limit at all locations per neighbourhood
traffic_best_2_worst = avg_percent.index
print(traffic_best_2_worst)

# Here is an array of pittsburgh neighborhoods best to worst


```
