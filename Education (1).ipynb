```python
import pandas as pd
import matplotlib.pyplot as plt
```

Importing Pandas and matplotlib.

```python
pd.set_option("display.max_rows", None)
ED = pd.read_csv("ed.csv",  index_col="Neighborhood")
```

Reading my data and displayign max rows.

```python
ED["Edu. Attainment: High School Graduate (2010)"] = (ED["Edu. Attainment: High School Graduate (2010)"].str.replace("%", "", regex=False))
ED["Edu. Attainment: High School Graduate (2010)"] = pd.to_numeric(ED["Edu. Attainment: High School Graduate (2010)"], errors="coerce")
```

Sorting my data then orginizing so the top is the bets and bottom is the worst.

```python
print(ED[["Edu. Attainment: High School Graduate (2010)"]] .sort_values("Edu. Attainment: High School Graduate (2010)", ascending=False) .head(10))
print(ED[["Edu. Attainment: High School Graduate (2010)"]] .sort_values("Edu. Attainment: High School Graduate (2010)", ascending=False) .plot())
```

Printing my Data in a more readable way.
