## Metric:

Crime rate

## Why is this useful?

I figured that finding the crime rate of each neighborhood would be a useful metric in determining the best neighborhood. When you're looking at areas to see which is best to live in, safety should definitely be a concern, and that's where the crime rate metric comes into play.

Keep in mind that Crime rates are measured by quantifying the number of reported crimes relative to the population of a specific area, typically expressed as offenses per 100,000 people.

```python
#importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#loading/reading the dataset and pre-formatting graph
read = pd.read_csv("arrests-for-major-crimes-1972.csv")
read = read.sort_values("overall_crime_rate", ascending = False)
fig, ax = plt.subplots(figsize=(10,9))
ax.barh(read["neighborhood"], read["overall_crime_rate"])

#labeling the graph
ax.set_xlabel("Crime Rate")
ax.tick_params(axis='y', labelsize=7)
ax.set_title("Crime in Pittsburgh Neighborhoods")

#displaying the data
plt.tight_layout()
plt.show()
```
